```python
import os
import kagglehub
import pandas as pd
```

```python
# Download latest version

# https://www.kaggle.com/datasets/bahramjannesarr/goodreads-book-datasets-10m/data
# https://github.com/BahramJannesar/GoodreadsBookDataset?tab=readme-ov-file

path = kagglehub.dataset_download("bahramjannesarr/goodreads-book-datasets-10m")

print("Path to dataset files:", path)
```

```python
book_csv_path = "..\data"
book_csv_list = os.listdir(book_csv_path)[:-1]
book_csv_path_list = [os.path.join(book_csv_path, book_csv) for book_csv in book_csv_list]

print(book_csv_path_list)
```

```python
dataframes = []
for path, name in zip(book_csv_path_list, book_csv_list):
    df = pd.read_csv(path)
    dataframes.append(df)

df = pd.concat(dataframes)

df.to_csv('goodreads.csv', index = False) #je ne sais pas si cette ligne marche car c'est 10 million de lignes
```
